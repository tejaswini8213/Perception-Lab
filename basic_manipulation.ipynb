```python
import cv2
from google.colab.patches import cv2_imshow
```

GRAYSCALE CONVERSION

```python
img = cv2.imread('image.jpg')
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY )
cv2_imshow(img_grayscale)
```

RESIZING

```python
(h, w) = img.shape[:2]
new_width = 400
aspect_ratio = h / w
new_height = int(new_width * aspect_ratio)
resized_image = cv2.resize(img, (new_width, new_height))
```

```python
cv2_imshow(resized_image)
```

ROTATION

```python
height, width = resized_image.shape[:2]
# get the center coordinates of the image to create the 2D rotation matrix
center = (width/2, height/2)
# using cv2.getRotationMatrix2D() to get the rotation matrix
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=45, scale=1)
# rotate the image using cv2.warpAffine
rotated_image = cv2.warpAffine(src=resized_image, M=rotate_matrix, dsize=(width, height))
cv2_imshow(rotated_image)
```

```python
(b_channel, g_channel, r_channel) = cv2.split(resized_image)
cv2_imshow(b_channel)
cv2_imshow(g_channel)
cv2_imshow(r_channel)
```

```python
print(img.shape) # Print image shape
```

```python
cropped_image = img[80:280, 150:330]
cv2_imshow(cropped_image)
```

```python
# Blur the image for better edge detection
img_gs = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(img_gs, (3,3), 0)
edges = cv2.Canny(image=resized_image, threshold1=100, threshold2=200) # Canny Edge Detection
# Display Canny Edge Detection Image
cv2_imshow(edges)
```

SOBEL X

```python
#Detection on the X axis
sobelx = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5) # Sobel Edge
cv2_imshow(sobelx)
```

SOBEL Y

```python
#Detection on the Y axis
sobely = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5) # Sobel Edge
cv2_imshow(sobely)
```

COMBINED XY SOBEL

```python
sobelxy = cv2.Sobel(src=img_blur, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)
cv2_imshow(sobelxy)
```
